```python
import random

# --- RizzAI: Professional + Sassy AI Chatbot ---

def get_response(user_input):
    user_input = user_input.lower()
    responses = {
        "hello": ["Hey there!", "Hello!", "Hi! How’s it going?"],
        "how are you": ["Doing great, thanks for asking.", "Efficient as always.", "Functioning at 99.9% efficiency."],
        "what is your name": ["I'm RizzAI — your AI assistant.", "They call me RizzAI, your friendly digital brain."],
        "bye": ["Goodbye!", "Catch you later.", "See you soon."],
        "default": [
            "Interesting... go on.",
            "That’s one way to look at it.",
            "Noted.",
            "I don’t have data for that, but sounds fascinating."
        ]
    }

    for key in responses:
        if key in user_input:
            return random.choice(responses[key])
    return random.choice(responses["default"])


def main():
    print("RizzAI: Professional AI Assistant (type 'exit' to quit)")
    print("-" * 55)
    while True:
        user_input = input("You: ")
        if user_input.lower() in ["exit", "quit"]:
            print("RizzAI: Session ended. Have a productive day!")
            break
        print("RizzAI:", get_response(user_input))


if __name__ == "__main__":
    main()
```
